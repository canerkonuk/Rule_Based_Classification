##### Task 2: What are the average earnings in COUNTRY, SOURCE, SEX, AGE breakdown? #####

##### Task 3: Sort the output obtained in Task 2 according to PRICE. #####

```python
# To see the output in the previous question better, apply the sort_values â€‹â€‹method in descending order of PRICE.
# Save the output as agg_df.
agg_df = df.groupby(["COUNTRY", "SOURCE", "SEX", "AGE"]).agg({"PRICE": ["mean"]}).sort_values(("PRICE", "mean"), ascending=False)
agg_df
```

##### Task 4: Convert the names in the index to variable names for agg_df output. #####

```python
# All variables except PRICE in the output of question 3 are index names. Convert these names to variable names.
agg_df.reset_index(inplace=True)
agg_df
```

##### Task 5: Convert age variable to categorical variable and add it to agg_df. #####

```python
# Age sayÄ±sal deÄŸiÅŸkenini kategorik deÄŸiÅŸkene Ã§eviriniz.
# AralÄ±klarÄ± ikna edici ÅŸekilde oluÅŸturunuz.
# Ã–rneÄŸin: â€˜0_18', â€˜19_23', '24_30', '31_40', '41_70'

agg_df["CAT_AGE"]=pd.cut(agg_df["AGE"], [0, 18, 23, 30, 40, 70], labels=["0_18", "19_23", "24_30", "31_40", "41_70"])
agg_df
```

##### Task 6: Identify new level-based customers (personas). #####

```python
# Define new level-based customers (personas) and add them as variables to the dataset.
# Name of the new variable to be added: customers_level_based
# You need to create the customers_level_based variable by combining the observations from the output you obtained in the previous task.
# Attention! After creating customers_level_based values â€‹â€‹with list comprehension, these values â€‹â€‹need to be deduplicated.
# For example, more than one of the following expressions: USA_ANDROID_MALE_0_18. It is necessary to take them to groupby and get the price averages.

agg_df["customers_level_based"] = [(k+"_"+l+"_"+m+"_"+n).upper() for k, l, m, n in zip(agg_df["COUNTRY"], agg_df["SOURCE"], agg_df["SEX"], agg_df["CAT_AGE"])]
agg_df
```

```python
#Creating new dataframe for segmentation and cleaning PRICE column.
segmentation_df = agg_df.groupby("customers_level_based").agg({("PRICE","mean"): ["mean"]})
segmentation_df = segmentation_df.droplevel(axis=1, level=[1,2]).reset_index()
segmentation_df
```

##### Task 7: Segment new customers (personas). #####

```python
# Divide new customers (Example: USA_ANDROID_MALE_0_18) into 4 segments according to PRICE.
# Add the segments to segmentation_df as a variable with the SEGMENT column.
# Describe segments (Group by segments and get price mean, max, sum).

segmentation_df["SEGMENT"]=pd.qcut(segmentation_df["PRICE"], q=4, labels=["D", "C", "B", "A"])
segmentation_df.groupby("SEGMENT").agg({"PRICE": ["mean", "max", "sum"]})
```
